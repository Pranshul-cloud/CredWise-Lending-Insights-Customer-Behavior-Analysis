```python

years = [2012, 2013, 2014, 2015]
loan_amounts = [38696500.00, 148637450.00, 67329425.00, 23352775.00]
installments = [429.58, 456.13, 445.28, 439.66]
interest_rates = [14.13, 14.65, 13.71, 12.99]


df = pd.DataFrame({
    'Year': years,
    'Loan_Amount': loan_amounts,
    'Installment': installments,
    'Interest_Rate': interest_rates
})

```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.barplot(data =df,x='Year',y='Loan_Amount',color='#2E86C1',edgecolor='grey')
plt.title('Total Loan Amount per Year',fontweight = 'bold')
plt.xlabel('Year',size=15,color='grey')
plt.ylabel('Loan Amount',size=15,color='grey')

plt.subplot(1,3,2)
sns.lineplot(data =df,x='Year',y='Interest_Rate',marker='o',color='#2E86C1')
plt.title('Avg Interest(%) per year',fontweight='bold')
plt.xlabel('Year',size=15,color='grey')
plt.ylabel('Avg Interest rate',size=15,color='grey')

plt.subplot(1,3,3)
sns.barplot(data =df,x='Year',y='Installment',color='#2E86C1',edgecolor='grey')
plt.title('Avg Monthly Installment per year',fontweight='bold')
plt.xlabel('Year',size=15,color='grey')
plt.ylabel('Avg Installment amount',size=15)

plt.tight_layout()
plt.savefig(r'C:\Users\ritul\Desktop\Pranshul Docs\FINTECH_PROJECT\my_plot.png') 
plt.show()
```

```python
Purpose =['Car','CC','Dept_Concolidation','Home_improvement','House','Major_purchase','Medical','Moving','Other','Renewable_energy','Small_business','Vacation','Wedding']
Annual_income= [242999,38907238,95658124,12139755,661474,3767994,1470582,893754,8115839,93924,2037141,788034,477034]
Average_Current_Balance= [3383,1604,1581,2113,1331,1316,1565,1293,1666,800,1661,1524,878]

df = pd.DataFrame({
    'Purpose':Purpose,
    'Annual_income':Annual_income,
    'Average_Current_Balance':Average_Current_Balance
})
```

```python
df["Annual_income_norm"] = df["Annual_income"] / df["Annual_income"].max()
df["Balance_norm"] = df["Average_Current_Balance"] / df["Average_Current_Balance"].max()

plt.figure(figsize=(10,5))
sns.barplot(x="Purpose", y="Annual_income_norm", data=df, color="skyblue", alpha=0.7, label="Annual Income (scaled)")
sns.lineplot(x="Purpose", y="Balance_norm", data=df, color="red", marker="o", label="Avg Current Balance (scaled)")

plt.xticks(rotation=45)
plt.ylabel("Scaled Values (0-1)",size=13)
plt.xlabel('Purpose',size=13)
plt.title("Annual Income vs Average Current Balance by Purpose", fontsize=14, fontweight="bold")
plt.legend()
plt.tight_layout()
plt.show()
```
